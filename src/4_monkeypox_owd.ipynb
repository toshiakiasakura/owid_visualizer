```python
df = pd.read_csv("../dt_our_world/owid-monkeypox-data.csv", parse_dates=["date"])
```

```python
df.location.unique()
```

```python
cnts = ["World", 
        "Japan", "United States", "United Kingdom", "France", "Germany", 
        "China", "Brazil", "Nigeria", "Congo"
       ]
```

```python
with cplt.Single() as p:
    for c in cnts:
        dfM = df[ df.location == c]
        p.ax.plot(dfM.date, dfM.new_cases_smoothed, label=c)
    plt.legend()
    
cnts.pop(0)
with cplt.Single() as p:
    for c in cnts:
        dfM = df[ df.location == c]
        p.ax.plot(dfM.date, dfM.new_cases_smoothed, label=c)
    plt.legend()
    
with cplt.Single() as p:
    for c in cnts:
        dfM = df[ df.location == c]
        p.ax.plot(dfM.date, dfM.new_cases_smoothed_per_million, label=c)
    plt.legend()
```
